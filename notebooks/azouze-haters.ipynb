```python
# ==========================================
# 1. SETUP AND IMPORTS
# ==========================================
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, BatchNormalization, Activation, Bidirectional, LSTM, Dense, Dropout, GlobalAveragePooling1D
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import os

# Check for GPU
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))

# ==========================================
# 2. DATA LOADING
# ==========================================
# Defining paths based on the 'mit-chb-processed' dataset structure
DATASET_PATH = '/kaggle/input/mit-chb-processed/'

print("Loading dataset...")
# Load the pre-processed numpy arrays
# X shape is typically (Samples, Channels, Time) -> (N, 18, 2048)
X = np.load(os.path.join(DATASET_PATH, 'signal_samples.npy'))
y = np.load(os.path.join(DATASET_PATH, 'is_sz.npy'))

print(f"Original X shape: {X.shape}")
print(f"Original y shape: {y.shape}")

# ==========================================
# 3. DATA PREPROCESSING
# ==========================================
# The model expects input shape (Batch, Time, Channels) for Conv1D
# Current shape is (Batch, Channels, Time) -> We need to transpose
X = np.transpose(X, (0, 2, 1))
print(f"Transposed X shape: {X.shape}") # Should be (N, 2048, 18)

# Binary Labels: The dataset might provide ratios (floats). We convert to binary (0 or 1).
# Usually, if ratio > 0 it implies seizure activity, but a threshold of 0.5 is safer.
# However, for detection, often any seizure content counts.
y_binary = (y > 0.5).astype(int)

# Split into Training and Test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.2, random_state=42, stratify=y_binary)

# Free up memory
del X, y

# Standardization (Optional but recommended for Neural Networks)
# We flatten, fit scaler, then reshape back.
# Note: For EEG, scaling per-sample or per-channel is often done. 
# Here we scale across the training set to preserve relative amplitude differences.
samples, time_steps, channels = X_train.shape
X_train_flat = X_train.reshape(-1, channels)
X_test_flat = X_test.reshape(-1, channels)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train_flat).reshape(samples, time_steps, channels)
X_test = scaler.transform(X_test_flat).reshape(X_test.shape[0], time_steps, channels)

print(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")

# ==========================================
# 4. MODEL ARCHITECTURE (Hybrid 1D-CNN + Bi-LSTM)
# ==========================================
# This architecture is based on high-performing models referenced in 2025 systematic reviews.
# It uses 1D Convolutions to extract high-frequency features and Bi-LSTM for temporal progression.

def build_hybrid_model(input_shape):
    inputs = Input(shape=input_shape)

    # --- Feature Extraction (CNN Block) ---
    # Block 1
    x = Conv1D(filters=64, kernel_size=5, padding='same')(inputs)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling1D(pool_size=2)(x)
    x = Dropout(0.3)(x)

    # Block 2
    x = Conv1D(filters=128, kernel_size=3, padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling1D(pool_size=2)(x)
    x = Dropout(0.3)(x)

    # Block 3
    x = Conv1D(filters=256, kernel_size=3, padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling1D(pool_size=2)(x)
    x = Dropout(0.3)(x)

    # --- Temporal Dynamics (RNN Block) ---
    # Bidirectional LSTM to capture seizure onset context from both directions
    x = Bidirectional(LSTM(128, return_sequences=True))(x)
    x = Dropout(0.4)(x)
    
    # Attention-like mechanism or Global Pooling
    # Global Average Pooling effectively summarizes the time dimension
    x = GlobalAveragePooling1D()(x)

    # --- Classification Head ---
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.5)(x)
    outputs = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=inputs, outputs=outputs)
    return model

input_shape = (X_train.shape[1], X_train.shape[2]) # (2048, 18)
model = build_hybrid_model(input_shape)

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy', tf.keras.metrics.AUC(name='auc')])

model.summary()

# ==========================================
# 5. TRAINING
# ==========================================
# Callbacks for best practices
checkpoint = ModelCheckpoint('best_model.keras', monitor='val_accuracy', save_best_only=True, mode='max', verbose=1)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=0.00001, verbose=1)
early_stop = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True, verbose=1)

history = model.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=50, # Adjust based on time limits, usually converges around 30-40
    batch_size=32,
    callbacks=[checkpoint, reduce_lr, early_stop],
    verbose=1
)

# ==========================================
# 6. EVALUATION AND METRICS
# ==========================================
# Load best model
model.load_weights('best_model.keras')

# Predict
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

# Metrics
acc = accuracy_score(y_test, y_pred)
print(f"\nTest Accuracy: {acc:.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['Non-Seizure', 'Seizure']))

# Confusion Matrix Visualization
import seaborn as sns
plt.figure(figsize=(6, 5))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Non-Seizure', 'Seizure'], yticklabels=['Non-Seizure', 'Seizure'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Training Curves
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Val')
plt.title('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Val')
plt.title('Loss')
plt.legend()
plt.show()# ==========================================
# 1. SETUP AND IMPORTS
# ==========================================
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, BatchNormalization, Activation, Bidirectional, LSTM, Dense, Dropout, GlobalAveragePooling1D
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import os

# Check for GPU
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))

# ==========================================
# 2. DATA LOADING
# ==========================================
# Defining paths based on the 'mit-chb-processed' dataset structure
DATASET_PATH = '/kaggle/input/mit-chb-processed/'

print("Loading dataset...")
# Load the pre-processed numpy arrays
# X shape is typically (Samples, Channels, Time) -> (N, 18, 2048)
X = np.load(os.path.join(DATASET_PATH, 'signal_samples.npy'))
y = np.load(os.path.join(DATASET_PATH, 'is_sz.npy'))

print(f"Original X shape: {X.shape}")
print(f"Original y shape: {y.shape}")

# ==========================================
# 3. DATA PREPROCESSING
# ==========================================
# The model expects input shape (Batch, Time, Channels) for Conv1D
# Current shape is (Batch, Channels, Time) -> We need to transpose
X = np.transpose(X, (0, 2, 1))
print(f"Transposed X shape: {X.shape}") # Should be (N, 2048, 18)

# Binary Labels: The dataset might provide ratios (floats). We convert to binary (0 or 1).
# Usually, if ratio > 0 it implies seizure activity, but a threshold of 0.5 is safer.
# However, for detection, often any seizure content counts.
y_binary = (y > 0.5).astype(int)

# Split into Training and Test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.2, random_state=42, stratify=y_binary)

# Free up memory
del X, y

# Standardization (Optional but recommended for Neural Networks)
# We flatten, fit scaler, then reshape back.
# Note: For EEG, scaling per-sample or per-channel is often done. 
# Here we scale across the training set to preserve relative amplitude differences.
samples, time_steps, channels = X_train.shape
X_train_flat = X_train.reshape(-1, channels)
X_test_flat = X_test.reshape(-1, channels)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train_flat).reshape(samples, time_steps, channels)
X_test = scaler.transform(X_test_flat).reshape(X_test.shape[0], time_steps, channels)

print(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")

# ==========================================
# 4. MODEL ARCHITECTURE (Hybrid 1D-CNN + Bi-LSTM)
# ==========================================
# This architecture is based on high-performing models referenced in 2025 systematic reviews.
# It uses 1D Convolutions to extract high-frequency features and Bi-LSTM for temporal progression.

def build_hybrid_model(input_shape):
    inputs = Input(shape=input_shape)

    # --- Feature Extraction (CNN Block) ---
    # Block 1
    x = Conv1D(filters=64, kernel_size=5, padding='same')(inputs)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling1D(pool_size=2)(x)
    x = Dropout(0.3)(x)

    # Block 2
    x = Conv1D(filters=128, kernel_size=3, padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling1D(pool_size=2)(x)
    x = Dropout(0.3)(x)

    # Block 3
    x = Conv1D(filters=256, kernel_size=3, padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling1D(pool_size=2)(x)
    x = Dropout(0.3)(x)

    # --- Temporal Dynamics (RNN Block) ---
    # Bidirectional LSTM to capture seizure onset context from both directions
    x = Bidirectional(LSTM(128, return_sequences=True))(x)
    x = Dropout(0.4)(x)
    
    # Attention-like mechanism or Global Pooling
    # Global Average Pooling effectively summarizes the time dimension
    x = GlobalAveragePooling1D()(x)

    # --- Classification Head ---
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.5)(x)
    outputs = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=inputs, outputs=outputs)
    return model

input_shape = (X_train.shape[1], X_train.shape[2]) # (2048, 18)
model = build_hybrid_model(input_shape)

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy', tf.keras.metrics.AUC(name='auc')])

model.summary()

# ==========================================
# 5. TRAINING
# ==========================================
# Callbacks for best practices
checkpoint = ModelCheckpoint('best_model.keras', monitor='val_accuracy', save_best_only=True, mode='max', verbose=1)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=0.00001, verbose=1)
early_stop = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True, verbose=1)

history = model.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=50, # Adjust based on time limits, usually converges around 30-40
    batch_size=32,
    callbacks=[checkpoint, reduce_lr, early_stop],
    verbose=1
)

# ==========================================
# 6. EVALUATION AND METRICS
# ==========================================
# Load best model
model.load_weights('best_model.keras')

# Predict
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

# Metrics
acc = accuracy_score(y_test, y_pred)
print(f"\nTest Accuracy: {acc:.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['Non-Seizure', 'Seizure']))

# Confusion Matrix Visualization
import seaborn as sns
plt.figure(figsize=(6, 5))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Non-Seizure', 'Seizure'], yticklabels=['Non-Seizure', 'Seizure'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Training Curves
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Val')
plt.title('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Val')
plt.title('Loss')
plt.legend()
plt.show()
```
